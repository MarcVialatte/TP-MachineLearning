## Exercice 3 - Classification

**Question 1**

Faire la visualisation de données du dataset (quelles types de données, données manquantes, taille
des images, répartition des classes, etc...)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

# Charger les données MNIST
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Visualiser quelques images
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(train_images[i], cmap='gray')
    plt.title(f"Label: {train_labels[i]}")
    plt.axis('off')
plt.show()

# Taille des images
print(f"Taille d'une image: {train_images[0].shape}")

# Répartition des classes
unique, counts = np.unique(train_labels, return_counts=True)
class_distribution = dict(zip(unique, counts))
print(f"Répartition des classes dans le jeu d'entraînement: {class_distribution}")

# Convertir les images en vecteurs
train_images_flat = train_images.reshape((len(train_images), -1))
test_images_flat = test_images.reshape((len(test_images), -1))

# Créer des DataFrames pour les jeux d'entraînement et de test
df_train = pd.DataFrame(train_images_flat)
df_test = pd.DataFrame(test_images_flat)

# Vérifier les données manquantes
missing_train = df_train.isnull().sum().sum()
missing_test = df_test.isnull().sum().sum()

print(f"Données manquantes dans le jeu d'entraînement : {missing_train}")
print(f"Données manquantes dans le jeu de test : {missing_test}")

```

**Question 2**

Normaliser le dataset. Expliquer ce que cela veut dire

```python
# Normalisation
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
```

**Question 3**

Séparer le jeu de données en train et test et faites un K-Nearest Neighbors. Tester plusieurs distances

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Aplatir les images en vecteurs
train_images_flat = train_images.reshape((len(train_images), -1))
test_images_flat = test_images.reshape((len(test_images), -1))

# Séparation en ensembles d'entraînement et de test
X_train, X_val, y_train, y_val = train_test_split(train_images_flat, train_labels, test_size=0.2, random_state=42)

# K-Nearest Neighbors avec différentes distances
distances = ['euclidean', 'manhattan', 'chebyshev']
for distance in distances:
    knn = KNeighborsClassifier(n_neighbors=3, metric=distance)
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_val)
    accuracy = metrics.accuracy_score(y_val, y_pred)
    print(f"Distance: {distance}, Accuracy: {accuracy}")
```

**Question 4**

Améliorer votre modèle (trouver les meilleurs paramètres) et essayer d’atteindre plus de 90% d'accuracy. Afficher aussi la Recall et le F1-score.

```python
# Trouver les meilleurs paramètres
best_accuracy = 0
best_params = {}

for distance in distances:
    for k in range(3, 10):
        knn = KNeighborsClassifier(n_neighbors=k, metric=distance)
        knn.fit(X_train, y_train)

        y_pred = knn.predict(X_val)
        accuracy = metrics.accuracy_score(y_val, y_pred)

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'distance': distance, 'k': k}

# Meilleurs paramètres
print(f"Meilleurs paramètres: {best_params}")

# Entraîner le modèle avec les meilleurs paramètres
knn_best = KNeighborsClassifier(n_neighbors=best_params['k'], metric=best_params['distance'])
knn_best.fit(X_train, y_train)

# Prédiction sur le jeu de test
y_pred_test = knn_best.predict(test_images_flat)

# Calculer et afficher l'accuracy, le recall et le F1-score
accuracy_test = metrics.accuracy_score(test_labels, y_pred_test)
recall_test = metrics.recall_score(test_labels, y_pred_test, average='weighted')
f1_test = metrics.f1_score(test_labels, y_pred_test, average='weighted')

print(f"Accuracy sur le jeu de test: {accuracy_test}")
print(f"Recall sur le jeu de test: {recall_test}")
print(f"F1-score sur le jeu de test: {f1_test}")
```
